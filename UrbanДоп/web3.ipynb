<h3 style="text-align: center;">Именование&nbsp;функций</h3>

<p>Имена функциям назначаются точно так же, как переменным. Имя функции должно быть достаточно описательным, чтобы любой читающий ваш код, мог догадаться, что именно делает функция.</p>

<p>Python и тут требует соблюдения тех же правил, что при именовании переменных:</p>

<ol>
	<li>
	<p>в имени функции используются&nbsp;только латинские буквы a-z, A-Z, цифры и символ нижнего подчеркивания (_);</p>
	</li>
	<li>
	<p>имя функции не может начинаться с цифры;</p>
	</li>
	<li>
	<p>имя&nbsp;функции по возможности должно отражать ее&nbsp;назначение;</p>
	</li>
	<li>
	<p>символы верхнего и нижнего регистра различаются.</p>
	</li>
</ol>

<p>Поскольку функции<strong> выполняют действия</strong>, большинство программистов предпочитает в именах функций <strong>использовать</strong> <strong>глаголы</strong>. Например:</p>

<ul>
	<li>функцию, которая <strong>рисует прямоугольник</strong>, можно назвать&nbsp;<code>draw_box()</code>;</li>

```python
# объявление функции
def draw_box():
    print('*' * 10)
    for _ in range(12):
        print('*        *')
    print('*' * 10)


# основная программа
draw_box()  # вызов функции
```

```python
# объявление функции
def draw_box(height, width):    # функция принимает два параметра
    print('*' * height)
    for _ in range(width):
        print('*        *')
    print('*' * height)


# основная программа
draw_box(10, 12)  # вызов функции
```

<p>Для функции <code>draw_box(height, width)</code>:</p>

<p>параметрами являются переменные <code>height</code> и <code>width</code>.</p>

<p>В момент вызова функции <code>draw_box(height, width)</code>:</p>

<p>Когда аргумент передается в функцию, параметрическая переменная функции будет ссылаться на значение этого аргумента. Однако любые изменения, которые вносятся в параметрическую переменную, не будут влиять на аргумент.</p>

<h2 style="text-align: center;">Примечание</h2>

<p><strong>Примечание 1.</strong> Функции в Python&nbsp;могут принимать сколько угодно параметров.</p>

<p><strong>Примечание 2.</strong> Иногда вместо параметров и аргументов говорят о <strong>формальных параметрах</strong>&nbsp;и <strong>фактических параметрах</strong>.&nbsp;Формальные параметры – переменные, которые мы пишем при описании функции. Фактические параметры – то, что реально подставляется при вызове функции.</p>


```python
# объявление функции
def draw_box(height, width):
    print('*' * height)
    for _ in range(width):
        print('*', ' ' * (height - 4), '*')
    print('*' * height)


# основная программа
draw_box(15, 10)
draw_box(5, 4)
draw_box(4, 5)
```

<h2 style="text-align: center;">Локальные переменные</h2>

<p><strong>Локальными</strong>&nbsp;называются переменные, объявленные внутри функции и доступные только ей самой. Программный код за пределами функции&nbsp;к ним&nbsp;доступа не имеет.</p>

<p>Если программный код одной функции попытается обратиться к локальной переменной, принадлежащей другой функции, произойдёт ошибка.</p>

<p>Локальные переменные скрыты от других функций, поэтому другие функции могут иметь собственные локальные переменные с тем же именем. 


<h2 style="text-align: center;">Область видимости переменной</h2>

<p>Область видимости переменной –&nbsp;часть программы, в которой можно к ней обращаться (та функция, где она создана). Переменная видима программному коду&nbsp;только в области её видимости. Никакая инструкция за пределами функции не может обращаться к такой переменной.</p>


<h2 style="text-align: center;">Область видимости параметрической переменной</h2>

<p>Область&nbsp;видимости параметрической переменной — функция, в которой этот параметр используется. К параметрической переменной имеет доступ весь программный код этой&nbsp;функции.</p>

<p>Рассмотрим уже известную нам функцию:</p>

<pre><code class="language-python hljs" data-highlighted="yes"><span class="hljs-keyword">def</span> <span class="hljs-title function_">draw_box</span>(<span class="hljs-params">height, width</span>):
    <span class="hljs-keyword">for</span> i <span class="hljs-keyword">in</span> <span class="hljs-built_in">range</span>(height):
        <span class="hljs-built_in">print</span>(<span class="hljs-string">'*'</span> * width)</code></pre>

<p>Параметрические переменные тут <code>height</code>, <code>width</code>. Внутри функции объявляется одна локальная переменная <code>i</code>.</p>

<h2 style="text-align: center;">Примечания</h2>

<p><strong>Примечание 1.</strong> Параметрическая переменная тоже локальная.</p>

<p><strong>Примечание 2.</strong>&nbsp;Память для локальных&nbsp;переменных выделяется на время исполнения данной функции&nbsp;в специальной области, называемой <a href="https://ru.wikipedia.org/wiki/%D0%A1%D1%82%D0%B5%D0%BA" rel="noopener noreferrer nofollow" target="_blank">стеком</a>. При завершении работы функции&nbsp;память освобождается,&nbsp;внутренние результаты работы функции&nbsp;не сохраняются от одного обращения к другому.</p>


```python
def print_nums():
    a = 100
    b = 200
    print(a + b)

print_nums()
print(a)
```

```python
def print_nums():
    a = 100
    b = 200
    print(a + b)


def print_foreign_num():
    print(a + b)


print_nums()

print_foreign_num()
```

<h2 style="text-align: center;">Глобальные переменные – зло</h2>

<p>Большинство программистов согласны, что следует ограничить использование глобальных переменных либо не использовать их вообще. Причины следующие.</p>

<ul>
	<li><strong>Глобальные переменные затрудняют отладку программы.</strong> Значение глобальной переменной может быть изменено любой инструкцией в программном файле. Если обнаружится, что в глобальной переменной хранится неверное значение, то придется отыскать все инструкции, которые к ней обращаются, чтобы определить, откуда поступает плохое значение. В программе с тысячами строк кода сделать это непросто.</li>
	<li><strong>Функции, использующие&nbsp;глобальные переменные, обычно зависят от этих переменных.</strong> Если возникнет необходимость применить такую функцию в другой программе, скорее всего придется эту функцию перепроектировать, чтобы она не опиралась на глобальную переменную.</li>
	<li><strong>Глобальные переменные затрудняют понимание программы.</strong> Глобальная переменная может быть модифицирована любой инструкцией в программе. При необходимости разобраться в какой-то части программы, использующей глобальную переменную, придется узнать обо всех других частях программы, обращающихся к этой глобальной переменной.</li>
</ul>

<p><img alt="" height="49" src="https://ucarecdn.com/bc151807-8987-423e-9b55-bea0ae2e654f/" style="float: left;" width="44">В большинстве случаев следует создавать переменные локально и передавать их в качестве аргументов в функции, которым нужно к ним обратиться.</p>

<h2 style="text-align: center;">Глобальные константы</h2>

<p>Хотя следует избегать использования глобальных переменных, в программе допускается применение глобальных констант. Глобальная константа – глобальное имя, ссылающееся на неизменное значение. Поскольку значение глобальной константы не может быть изменено во время исполнения программы, можно не беспокоиться о потенциальных опасностях, обычно связанных с использованием глобальных переменных.</p>

<p>Несмотря на то, что язык Python не позволяет создавать настоящие глобальные константы, их можно имитировать при помощи глобальных переменных. Если глобальная переменная не объявляется с использованием ключевого слова <code>global</code> внутри функции, то внутри этой функции вы никак не сможете изменить значение этой глобальной переменной.&nbsp;</p>

<p><img alt="" height="49" src="https://ucarecdn.com/bc151807-8987-423e-9b55-bea0ae2e654f/" style="float: left;" width="44">Математический модуль <code>math</code>&nbsp; определяет две глобальные&nbsp;переменные, <code>math.pi</code>&nbsp;и <code>math.e</code>, которым присвоены математические значения констант&nbsp;<em><a href="https://ru.wikipedia.org/wiki/%D0%9F%D0%B8_(%D1%87%D0%B8%D1%81%D0%BB%D0%BE)" rel="noopener noreferrer nofollow" target="_blank">π</a>&nbsp;</em>= <code>3.14159265</code>&nbsp;и <em><a href="https://ru.wikipedia.org/wiki/E_(%D1%87%D0%B8%D1%81%D0%BB%D0%BE)" rel="noopener noreferrer nofollow" target="_blank">e</a>&nbsp;</em>=&nbsp;<code>2.71828</code>.</p>

<h2 style="text-align: center;">Ключевое слово global</h2>

<p>Если нужно, чтобы инструкция внутри функции присваивала значение глобальной переменной, то требуется дополнительный шаг. В этом случае, глобальная переменная должна быть объявлена внутри функции.</p>

<h2 style="text-align: center;">Примечания</h2>

<p><strong>Примечание 1.</strong> При написании функций стоит указывать более важные параметры первыми.</p>

<p><strong>Примечание 2.</strong> Именованные аргументы часто используют&nbsp;вместе со&nbsp;значениями по&nbsp;умолчанию.</p>

<p><strong>Примечание 3.</strong> Именованные и&nbsp;позиционные аргументы не&nbsp;всегда хорошо ладят друг с&nbsp;другом. При вызове функции позиционные аргументы должны обязательно идти перед именованными аргументами.</p>

<p><strong>Примечание 4.</strong> Параметр в программировании — принятый функцией аргумент. Термин «аргумент» подразумевает, что конкретно и какой конкретной функции было передано, а параметр — в каком качестве функция применила это принятое. То есть вызывающий код передает аргумент в параметр, который определен в описании (заголовке) функции.</p>

<ul>
	<li><strong>P</strong>arameter&nbsp;→ <strong>P</strong>laceholder (заполнитель принадлежит имени функции и используется в теле функции);</li>
	<li><strong>A</strong>rgument&nbsp;→ <strong>A</strong>ctual value (фактическое значение, которое передается при вызове функции).</li>
</ul>

<p>Подробнее можно почитать <a href="https://stackoverflow.com/questions/156767/whats-the-difference-between-an-argument-and-a-parameter" rel="noopener noreferrer nofollow" target="_blank">тут</a>.</p>

<p><strong>Примечание 5.</strong> Отличная <a href="https://habr.com/ru/company/ruvds/blog/515678/" rel="noopener noreferrer nofollow" target="_blank">статья</a> на хабре про аргументы функций в Python.</p>


```python
def fancy(length, char1='-', char2='*'):
    return (char1 + char2) * length + char1


print(fancy(3))
print(fancy(3, '.'))
print(fancy(2, ':', '|'))
print(fancy(4, char2='#'))
print(fancy(char2='$', length=3))
```

```python
# Ошибка из-за отсутствия обязательного аргумента в параметре length

def fancy(length, char1='-', char2='*'):
    return (char1 + char2) * length + char1


print(fancy(char2='!'))
```

<p>В заголовке функции&nbsp;<code>my_func()</code> указан всего один параметр&nbsp;<code>args</code>, но&nbsp;со звездочкой перед ним. Звездочка в определении функции означает, что переменная (параметр) <code>args</code> получит в виде кортежа все аргументы, переданные в функцию при ее вызове от текущей позиции и до конца.</p>

<p>При описании функции можно использовать <strong>только один</strong> параметр&nbsp;помеченный звездочкой, причем располагаться он должен в конце списка параметров, иначе последующим параметрам не достанется значений.</p>



```python
def my_func(*args, num):
    print(f"{args: }")
    print(f"{num = }")


my_func(1, 2, 3, 4, 5)
```

```python
def my_func(num, *args):
    print(f"{args = }")
    print(f"{num = }")


my_func(1, 2, 3, 4, 5)
```

<p>Помеченный звездочкой параметр&nbsp;<code>*args</code>&nbsp;нормально переживает и отсутствие аргументов, в то время как&nbsp;позиционные&nbsp;параметры всегда обязательны.</p>



```python
def my_func(num, *args):
    print(f"{args = }")
    print(f"{num = }")


my_func(1)
```

<p>Обратите внимание: функция <code>my_func()</code>&nbsp;принимает несколько аргументов, но как минимум один аргумент должен быть передан обязательно. Этот первый аргумент станет значением переменной&nbsp;<code>num</code>, а остальные аргументы сохранятся в переменной&nbsp;<code>args</code>. Подобным образом можно делать любое&nbsp;количество&nbsp;обязательных<em> </em>аргументов.</p>

<h2 style="text-align: center;">Получение именованных аргументов в виде словаря</h2>

<p>Позиционные аргументы можно получать в виде&nbsp;<code>*args</code>, причём произвольное их количество. Такая возможность существует и для именованных аргументов. Только именованные аргументы получаются в виде словаря, что позволяет сохранить имена аргументов в ключах.</p>



```python
def my_func(**kwargs):
    print(type(kwargs))
    print(kwargs)


my_func()
my_func(a=1, b=2)
my_func(name='Roman', hobby='Developer')
```


<p><img alt="" height="49" src="https://ucarecdn.com/c6f2a334-adec-420d-87d3-d8a1226a9442/" style="float: left;" width="49">По соглашению параметр, получающий подобный словарь, принято называть&nbsp;<code>kwargs</code>&nbsp;(от словосочетания keyword arguments).&nbsp;Старайтесь придерживаться этого соглашения.</p>

<p>Параметр&nbsp;&nbsp;<code>**kwargs</code>&nbsp;пишется в самом конце,&nbsp;после последнего&nbsp;аргумента со значением по умолчанию. При этом функция может содержать и <code>*args</code> и <code>**kwargs</code> параметры.</p>



```python
def my_func(a, b, *args, name='Gvido', age=1000100, **kwargs):
    print(a, b)
    print(args)
    print(name, age)
    print(kwargs)

my_func(1, 1)
print()
my_func(1, 2, 'three', 'four', 'five', name='Roman', age=28, job='Developer', work_experience=50, language='Python')
```

<p>Не нужно пугаться, в реальном коде функции редко&nbsp;используют все эти возможности одновременно. Но понимать, как каждая отдельная форма объявления аргументов работает, и как такие формы&nbsp;сочетать — очень важно.</p>

<h2 style="text-align: center;">Примечания.</h2>

<p><strong>Примечание 1.</strong> Специальный синтаксис&nbsp;<code>*args</code>&nbsp;и&nbsp;<code>**kwargs</code>&nbsp;в определении функции позволяет передавать функции переменное количество позиционных и именованных&nbsp;аргументов. При этом <code>args</code> и <code>kwargs</code>&nbsp;<strong>просто имена.&nbsp;</strong>Вы не обязаны их использовать, можно выбрать любые, однако среди Python программистов приняты именно эти.</p>

<p><strong>Примечание 2.</strong> Вы можете использовать одновременно&nbsp;<code>*args</code>&nbsp;и&nbsp;<code>**kwargs</code>&nbsp;в одной строке для вызова функции. В этом случае&nbsp;<strong>порядок имеет значение</strong>. Как и аргументы, не являющиеся аргументами по умолчанию,&nbsp;<code>*args</code>&nbsp;должны предшествовать и аргументам по умолчанию,&nbsp;и&nbsp;<code>**kwargs</code>. Правильный порядок&nbsp;параметров:</p>

<ol>
	<li>позиционные аргументы,</li>
	<li><code>*args</code>&nbsp;аргументы,&nbsp;</li>
	<li><code>**kwargs</code>&nbsp;аргументы.</li>
</ol>

<pre><code class="language-python hljs" data-highlighted="yes"><span class="hljs-keyword">def</span>&nbsp;<span class="hljs-title function_">my_func</span>(<span class="hljs-params">a, b, *args, **kwargs</span>):</code></pre>


```python
def f(n=3):
    return n + 7


print(f(f(f())))
```

```python
def func(x, y, *args):
    return len(args)


print(func(10, 20, 30, 40, 50, 60))
```

```python
def func(*args):
    return max(args) + min(args)


print(func(10, 15, *[31, 42, 5, 1], *(17, 28, 19, 100), 13, 12))
```

<p>Напишите функцию <code>print_products()</code>, которая принимает&nbsp;произвольное количество аргументов&nbsp;и <strong>выводит</strong> список продуктов (любая&nbsp;непустая&nbsp;строка) по образцу:&nbsp;<code>&lt;номер продукта&gt;)&nbsp;&lt;название продукта&gt;</code> (нумерация продуктов&nbsp;начинается с единицы).&nbsp;Если среди&nbsp;переданных аргументов нет ни одного продукта,&nbsp;необходимо <strong>вывести</strong> текст <code>Нет продуктов</code>.</p>

```python
def print_products(*args):
    '''
        Функция для вывода списка продуктов из свалки аргументов
    '''
    groceries = [arg for arg in args if arg and isinstance(arg, str)]
    if groceries:
        [print(f'{n + 1}) {item}') for n, item in enumerate(groceries)]
    else:
        print('Нет продуктов')


print_products('Бананы', [1, 2], ('Stepik',), 'Яблоки', '', 'Макароны', 5, True)
```
